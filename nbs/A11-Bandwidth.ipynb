```python
import numpy as np
import matplotlib.pyplot as plt
from quests.tools.plotting import savefig
from quests.entropy import get_bandwidth

plt.style.use("paper")
```

```python
v = np.linspace(5, 40)
bw = get_bandwidth(v)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(v, bw)
ax.set_xlabel("Volume (Å$^3$/atom)")
ax.set_ylabel("Bandwidth (Å$^{-1}$)")

savefig(fig, "01-bandwidth.pdf", prefix="../figs")
plt.show()
```
