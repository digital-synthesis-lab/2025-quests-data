```python
r = np.linspace(0.1, 6, 51)
w = np.array([cutoff_fn(x, cutoff=5) for x in r])
```

```python
fig, ax = plt.subplots()

ax.plot(r, 1 / r, label="1 / r")
ax.plot(r, 1 / np.power(r, 3 / 2), label="1 / r$^{3/2}$")
ax.plot(r, w / r, label="w(r) / r")
ax.plot(r, w / np.power(r, 3 / 2), label="w(r) / r$^{3/2}$")

ax.set_xlim(0, 5)
ax.set_ylim(0, 2)

ax.set_ylabel("Alternative X$_1$")
ax.set_xlabel("r (Ã…)")

ax.legend(frameon=False)
# savefig(fig, "01-x-scaling.pdf")
plt.show()
```
